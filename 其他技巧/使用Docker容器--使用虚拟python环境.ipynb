1.使用Python Docker容器
解决的问题
我们在这里解决什么问题：

假设您要使用python版本3处理python项目，但不想在计算机上安装/升级python
您希望在一个使用和投射环境中检查一个python代码（以检查其与特定python版本的兼容性）。在此，您不想在计算机上安装多个python版本及其必需的依赖项。
您正在学习python，将python用于实验目的。
所有这些问题的解决方案是搬运工容器。多克尔容器可帮助您创建一个可抛出的虚拟环境，您可以在其中运行应用程序而无需在计算机上安装蟒或任何必需的依赖项。

运行一个Docker容器
先决条件：

Docker应该安装在您的机器上，如果没有，请从Docker站点下载并按照您的操作系统的说明进行操作。

命令：

docker run --it --name my-python-app -v "$(pwd)":/root/ python:3 /bin/bash

参数说明：

run：在您的机器上查找一个docker映像（如果不可用），从Docker Hub下载它，然后从该映像运行一个容器。

--it ：在异步Bash会话（/ bin / bash）中启动容器

--name ：您的容器名称

-v ：要从主机上装载到容器的卷

python：3：是图像名称，后跟标记（python 3.6.1版本当前标记为3）

2.使用Python虚拟环境
python环境问题
处理python项目的典型方法是使用pip安装所有必需的附件。但是替换安装有什么作用？以及所有其他安装在替换的python的站点位置中。这可行，但是存在以下问题：

该项目的环境不干净。即，如果您正在处理多个项目，则在同一位置安装一堆python蟒蛇。
如果您尝试冻结当前项目的需求，则生成的需求将是在另一位置安装的所有python的列表。
如果多个项目需要不同版本的python扩展，那么您会遇到麻烦。
如果您未以管理员身份登录，则要安装多个，必须切换用户（具有admin访问权限），因为pip在系统库中安装了python范围。
为了克服所有这些问题，使用了虚拟环境。

如何使用虚拟环境：
安装虚拟环境：
sudo pip install virtualenv
创建一个新的虚拟环境（例如venv）
virtualenv venv
激活虚拟环境
source venv/bin/activate
使用此虚拟环境在Requirements.txt中安装依赖项
(venv) $ pip install -r requirements.txt
您已准备好出发！！您可以检查安装的任何重叠的位置，它应该列出您的虚拟环境中的位置，例如：

(venv) $ pytest --version
这是从/venv/lib/python2.7/site-packages/pytest.pyc导入的pytest版本3.0.7，使用终端执行任何项目文件，例如：

(venv) $ pytest tests/
如果您使用的是pyCharm IDE，则应该（实际）使用此虚拟环境

从顶部选项->项目解释器->选择venv作为项目解释器。在拆分列表中，您只会看到从requirements.txt下载的副本。

完成之后，并想要一个新的虚拟环境，只需使用以下命令：

(venv) $ deactivate
这将终止当前的虚拟环境。

笔记
每次启动bash提示时，都必须激活虚拟环境。
